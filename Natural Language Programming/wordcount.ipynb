```python
import pandas as pd

import matplotlib.pyplot as plt
from wordcloud import WordCloud

import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer, PorterStemmer
!unzip /usr/share/nltk_data/corpora/wordnet.zip -d /usr/share/nltk_data/corpora/

import string
```

```python
filename = "/kaggle/input/review/Review - Review.csv"
file = pd.read_csv(filename, usecols = ["review"])
text = " ".join(file["review"])
```

```python
tokens = word_tokenize(text.lower())

stop_words = set(stopwords.words('english'))
stop_words.add("br")
```

```python
lemmatizer = WordNetLemmatizer()
stemmer = PorterStemmer()
filtered_tokens = [stemmer.stem(lemmatizer.lemmatize(word)) for word in tokens if word not in stop_words and word not in string.punctuation]

cleaned_text = ' '.join(filtered_tokens)
```

```python
wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='white', 
                stopwords = stop_words, 
                min_font_size = 10).generate(cleaned_text)

plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
plt.show()
```
